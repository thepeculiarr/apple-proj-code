```python
#Importing Relevant Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

#loading data
df_data = pd.read_csv('C:\\Users\\PECU\\Documents\\FoodBalanceSheets_E_Africa_NOFLAG.csv', encoding = "latin-1")
df_data.head(60)


#data description
df_data.describe()

#shape of data
df_data.shape

#Checking for missing values
df_data.isnull().sum()

#using groupby to count the sum of each unique value in the Y2014,Y2015,Y2016,Y2017,Y2018 columns
df_data.groupby('Y2014')['Y2014'].count()
df_data[['Y2014']] = df_data[['Y2014']].fillna(value = 'mfc')


df_data.groupby('Y2015')['Y2015'].count()
df_data[['Y2015']] = df_data[['Y2015']].fillna(value = 'mfc')


df_data.groupby('Y2016')['Y2016'].count()
df_data[['Y2016']] = df_data[['Y2016']].fillna(value = 'mfc')


df_data.groupby('Y2017')['Y2017'].count()
df_data[['Y2017']] = df_data[['Y2017']].fillna(value = 'mfc')


df_data.groupby('Y2018')['Y2018'].count()
df_data[['Y2018']] = df_data[['Y2018']].fillna(value = 'mfc')



#Checking if missing values have been filled
df_data.isnull().sum()


#Using sum() to sum the rows of each columns
df = df_data.sum()
print(df)


#Printing columns
print(df_data.columns)


#Knowing my data type
type(df)

#data information
df.info()

#knowing my data type2
type(df_data)

#data information2
df_data.info()

#performing a groupby on elements to using df_data
df_data.groupby(['Element','Area']).count()

#Performing a groupby on element
df_data.groupby(['Element']).count()

#Performing a groupby on Y2017 and Area
df_data.groupby(['Y2017','Area']).count()

#Performing a groupby on element code
df_data.groupby(['Element Code']).count()

#Performing a groupby on Element and Item
df_data.groupby(['Element','Item']).count()
```
